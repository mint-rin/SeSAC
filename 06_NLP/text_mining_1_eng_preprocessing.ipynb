# 데이터 수집
- 온라인 상에서 데이터 수집해서 실습
https://www.forbes.com/sites/adrianbridgwater/2019/04/15/what-drove-the-ai-renaissance/?ss=ai-big-data&sh=2f9b595a1f25


```python
import requests
from bs4 import BeautifulSoup
resp = requests.get('https://www.forbes.com/sites/adrianbridgwater/2019/04/15/what-drove-the-ai-renaissance/?ss=ai-big-data&sh=2f9b595a1f25')
bs = BeautifulSoup(resp.text, 'html.parser')
```

```python
div = bs.find('div', class_='article-body fs-article fs-responsive-text current-article')
p_tag = div.find_all('p')
content = ''

for i in p_tag:
  content+=i.text

print(content)
```

# 영문 토큰화
- punkt tokenizer 참고: https://www.nltk.org/_modules/nltk/tokenize/punkt.html


word_tokenize() : 마침표와 구두점(온점, 반점, 물음표, 느낌표, 세미콜론 등과 같은 기호)으로 구분하여 토큰화

```python
# word_tokenize
import nltk

# nltk punkt tokenizer download
# 사전 다운로드
nltk.download('punkt')

from nltk.tokenize import word_tokenize
token1 = word_tokenize(content)

print(token1)
```

## WordPunctTokenizer() : 알파벳이 아닌 문자를 구분하여 토큰화

```python
# WordPunctTokenizer(): 알파벳이 아닌 문자를 구분하여 토큰화
import nltk

from nltk.tokenize import WordPunctTokenizer
token2 = WordPunctTokenizer().tokenize(content)
print(token2)
```

## TreebankWordTokenizer() : 정규표현식에 기반한 토큰화

```python
# TreebankWordTokenizer() : 정규표현식에 기반한 토큰화

import nltk

from nltk.tokenize import TreebankWordTokenizer
token = TreebankWordTokenizer().tokenize(content)
print(token[:20])
```

# 영문 개체명 인식
http://www.nltk.org/api/nltk.chunk.html

```python
nltk.download('words')
nltk.download('maxent_ne_chunker')
```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

# 토큰화
token1 = word_tokenize('Barack Obama likes fried chicken very much')
print('token: ', token1)
```

```python
# pos-tag
taggedToken = pos_tag(token1)
print('pos-tag: ', taggedToken)
```

```python
# chunking
from nltk import ne_chunk
neToken = ne_chunk(taggedToken)
print(neToken)
```
